```python
import contextily as cx
import geopandas as gpd
import matplotlib.pyplot as plt
import os
import requests
import shapely
import zipfile
```

```python
zip_path = "/home/manda/Documents/Leyard_projects/gis_data/2024eclipse_shapefiles.zip"
extract_path = "/home/manda/Documents/Leyard_projects/gis_data/eclipse_2024_extracted"

# Create directory if it doesn't exist
os.makedirs(extract_path, exist_ok=True)

# Extract
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

```python
extract_path = "/home/manda/Documents/Leyard_projects/gis_data/eclipse_2024_extracted"

# Load shapefiles (high-res path and umbra)
path_gdf = gpd.read_file(os.path.join(extract_path, "upath_hi.shp"))
umbra_gdf = gpd.read_file(os.path.join(extract_path, "umbra_hi.shp"))
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15, 7)

path_gdf.plot(ax=ax, facecolor='#969696', edgecolor='none', alpha=0.5)
umbra_gdf.plot(ax=ax, facecolor='#252525', edgecolor='none')

plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15, 7)

path_gdf.plot(ax=ax, facecolor='#969696', edgecolor='none', alpha=0.5)
umbra_gdf.plot(ax=ax, facecolor='#252525', edgecolor='none')

cx.add_basemap(ax, crs=path_gdf.crs, source=cx.providers.OpenTopoMap)

plt.show() 

fig.savefig("eclipse_map.png", dpi=300, bbox_inches='tight')
```
